```python
import pencil as pc

ff=pc.read.var(var_file='var.dat',trimall=True)
#f0=pc.read.var(ivar=0,trimall=True)
dim=pc.read.dim()
par=pc.read.param()

fig, ((ax1,ax2,ax3,ax4),(ax5,ax6,ax7,ax8)) = plt.subplots(2,4,figsize=(8,4),sharex=True,sharey=True)

x=ff.x
z=ff.z

drho=ff.rho[:,0,:]#-f0.rho[:,0,:]
dux=ff.ux[:,0,:]#-f0.ux[:,0,:]
duy=ff.uy[:,0,:]#-f0.uy[:,0,:]
duz=ff.uz[:,0,:]#-f0.uz[:,0,:]

dnd=ff.nd[:,0,:]#-f0.nd[:,0,:]
dudx=ff.uud1x[:,0,:]#-f0.uud1x[:,0,:]
dudy=ff.uud1y[:,0,:]#-f0.uud1y[:,0,:]
dudz=ff.uud1z[:,0,:]#-f0.uud1z[:,0,:]

ax1.contourf(x,z,drho,256)
ax2.contourf(x,z,dux,256)
ax3.contourf(x,z,duy,256)
ax4.contourf(x,z,duz,256)

ax5.contourf(x,z,dnd,np.linspace(0,5,256))
ax6.contourf(x,z,dudx,256)
ax7.contourf(x,z,dudy,256)
ax8.contourf(x,z,dudz,256)

ax1.set_title('Gas Density')
ax2.set_title('Gas x-velocity')
ax3.set_title('Gas y-velocity')
ax4.set_title('Gas z-velocity')

ax5.set_title('Dust Density')
ax6.set_title('Dust x-velocity')
ax7.set_title('Dust y-velocity')
ax8.set_title('Dust z-velocity')

ax1.set_aspect('equal')
ax2.set_aspect('equal')
ax3.set_aspect('equal')
ax4.set_aspect('equal')
ax5.set_aspect('equal')
ax6.set_aspect('equal')
ax7.set_aspect('equal')
ax8.set_aspect('equal')


plt.tight_layout()


print(ff.t)
```
